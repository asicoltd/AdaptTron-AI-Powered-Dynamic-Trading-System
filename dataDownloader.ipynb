```python
import os
import zipfile
import requests
from datetime import datetime, timedelta

# Function to download the file if it doesn't already exist
def download_file(date_str, download_dir, interval):
    zip_filename = os.path.join(download_dir, f"BTCUSDT-{interval}-{date_str}.zip")
    
    # Check if the file already exists
    if os.path.exists(zip_filename):
        print(f"File {zip_filename} already exists. Skipping download.")
        return
    
    # If not, proceed to download
    url = f"https://data.binance.vision/data/spot/daily/klines/BTCUSDT/{interval}/BTCUSDT-{interval}-{date_str}.zip"
    print(f"Downloading {zip_filename}...")
    response = requests.get(url)
    
    if response.status_code == 200:
        # Save the zip file
        with open(zip_filename, 'wb') as file:
            file.write(response.content)
    else:
        print(f"Failed to download {url}, status code: {response.status_code}")

# Function to unzip all downloaded files
def unzip_files(download_dir):
    print(f"Extracting files in {download_dir}...")
    for file in os.listdir(download_dir):
        if file.endswith(".zip"):
            zip_path = os.path.join(download_dir, file)
            extract_folder = os.path.join(download_dir, file.replace(".zip", ""))
            
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(extract_folder)
            print(f"Extracted {zip_path}")

# Function to remove all downloaded ZIP files
def remove_zip_files(download_dir):
    print(f"Removing ZIP files from {download_dir}...")
    for file in os.listdir(download_dir):
        if file.endswith(".zip"):
            zip_path = os.path.join(download_dir, file)
            os.remove(zip_path)
            print(f"Deleted {zip_path}")

# Define the date range
start_date = datetime(2017, 4, 17)
end_date = datetime(2025, 2, 19)

# Create the directory for storing the files if it doesn't exist

# Download all ZIP files for both 15m and 1m intervals (One file per day for each interval)
current_date = start_date
while current_date <= end_date:
    date_str = current_date.strftime("%Y-%m-%d")
    # Download for 15m interval
    download_dir = "Data/15m"
    download_file(date_str, download_dir, "15m")
    # Download for 1m interval
    download_dir = "Data/1m"
    download_file(date_str, download_dir, "1m")
    # Download for 1m interval
    download_dir = "Data/4h"
    download_file(date_str, download_dir, "4h")
    current_date += timedelta(days=1)  # Increment by one day

# Unzip all downloaded files
unzip_files(download_dir)

# Remove all ZIP files
remove_zip_files(download_dir)
```
